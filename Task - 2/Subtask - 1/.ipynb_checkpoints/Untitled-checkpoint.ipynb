```python
from torch.utils.data import DataLoader,Dataset

class RoadDataset(Dataset):
    def __init__(self, inputs, labels, transform, label_transform):
        self.x = [transform(i) for i in inputs]
        self.y = [label_transform(labels[i],inputs[i]) for i in range(len(labels))]

    def __len__(self):
        return len(self.y)

    def __getitem(self,idx):
        return self.x[idx], self.y[idx]
        
    
```

```python
def input_transform(x:Image):
    x = x.resize((512,512))
    x = np.asarray(x,dtype=np.float16)/255.0
    return x

def label_transform(x:Image, y:Image):
    x = x.resize((512,512))
    x = np.asarray(x,dtype=np.float16)/255.0
    y = y.resize((512,512))
    y = np.asarray(y,dtype=np.float16)/255.0
    x_ = x.mean(dim=-1)
    x_ = np.where(x_ == 0, 0, 1)
    return y*x_
    
```
