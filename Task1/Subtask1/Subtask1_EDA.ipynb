```python
import pandas as pd
import numpy as np
dataset_path="./ranking_dataset.csv"
df=pd.read_csv(dataset_path)
df.head()
```

```python

input_df=df[[f"val_{x}" for x in range(10)]]
output_df=df[[f"rank_{x}" for x in range(10)]]

input_df
```

```python
#First, we find the maximum and minimum of the input as well as the number of unique values

print("Global maximum:",input_df.max().max())
print("Global minimum:", input_df.min().min())
print("Maximum Range in a sequence:\n",(input_df.max(axis=1)-input_df.min(axis=1)).abs().max())
print("num_unique:" ,input_df.stack().nunique())
```

```python
rows_with_dupes = input_df[input_df.nunique(axis=1) < len(input_df.columns)]
ranks_with_dupes= output_df[output_df.nunique(axis=1) < len(output_df.columns)]
ranks_of_dupes = output_df[input_df.nunique(axis=1) < len(input_df.columns)]
print(len(ranks_with_dupes))
print(len(rows_with_dupes))
```
