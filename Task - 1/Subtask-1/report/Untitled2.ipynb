```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
accuraciesMLP =[]
accuraciesRNN =[]
accuraciesLSTM =[]
lossesMLP = []
lossesRNN = []
lossesLSTM = []
with open("./LSTMPredictor/log.txt","r" ) as f:
    accuraciesLSTM = list(map(float,f.readlines()))
with open("./LSTMPredictor/log2.txt","r" ) as f:
    lossesLSTM = list(map(float,f.readlines()))
with open("./MLPPredictor/log.txt","r" ) as f:
    accuraciesMLP = list(map(float,f.readlines()))
with open("./MLPPredictor/log2.txt","r" ) as f:
    lossesMLP = list(map(float,f.readlines()))
with open("./RNNPredictor/log.txt","r" ) as f:
    accuraciesRNN = list(map(float,f.readlines()))
with open("./RNNPredictor/log2.txt","r" ) as f:
    lossesRNN = list(map(float,f.readlines()))
```

```python
plt.plot(accuraciesMLP,label = "MLP")
plt.plot(accuraciesRNN,label = "RNN")
plt.plot(accuraciesLSTM,label = "LSTM")
plt.legend()
```

```python
lossesMLP = np.array(lossesMLP).reshape(100,-1).sum(axis=-1)
lossesRNN = np.array(lossesRNN).reshape(100,-1).sum(axis=-1)
lossesLSTM = np.array(lossesLSTM).reshape(100,-1).sum(axis=-1)


plt.plot(lossesMLP,label = "MLP")
plt.plot(lossesRNN,label = "RNN")
plt.plot(lossesLSTM,label = "LSTM")
plt.legend()
```
